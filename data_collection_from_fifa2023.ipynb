```python
import requests
import re
from bs4 import BeautifulSoup

with open("players.csv", mode='w', encoding="utf-8") as fd:
        fd.write("name,position,age,pace(PAC),Shooting(SHO),Passing(PAS),Dribbling(DRI),Defending(DEF),Physical(PHY),Attacking_Work_Rate(AWR),Defensive_Work_Rate(DWR),Preferred_Foot,club,player_nationality,Player_height,Player_weight,birthday_of_player\n")
        for page in range(1,372):
            try :
                print(f"scraping data from page {page}")
                res = requests.get(f"https://www.fifplay.com/fifa-23/players?page={page}")
                soup = BeautifulSoup(res.text, "html.parser")
                player = soup.find_all("td", attrs={"class":"name"})
                sates = soup.find_all("div", attrs={"class":"stats-main"})
                rate = soup.find_all("div", attrs={"class":"stats"})
                for i in range(len(player)):
                    name_player= player[i].text.strip().split("\n")[0]
                    position_player = player[i].text.strip().split("\n")[1]
                    ii = i * 6
                    pace = int(sates[ii].text)
                    Shooting = int(sates[ii+1].text)
                    passing = int(sates[ii+2].text)
                    Dribbling =int(sates[ii+3].text)
                    Defending = int(sates[ii+4].text)
                    Physical = int(sates[ii+5].text)
                    j = i*4
                    Attacking_Work_Rate = rate[j+1].text
                    Defensive_Work_Rate = rate[j+2].text
                    Preferred_Foot = rate[j+3].text
                    player_link = player[i].a["href"]
                    res = requests.get(player_link)
                    soup = BeautifulSoup(res.text, "html.parser")
                    info_player = soup.find_all("span", attrs={"class":"alignright bold"})
                    club = info_player[2].text
                    player_nationality = info_player[4].text
                    Player_height_weight= info_player[5].text.split("/")
                    Player_height = int(Player_height_weight[0].strip(" cm"))
                    Player_weight = int(Player_height_weight[1].strip(" kg"))
                    age= int(info_player[0].text.split()[0])
                    birthday= info_player[0].text
                    birthday_of_player = re.search(r'\((.*?)\)', birthday).group(1)
                    fd.write(f"{name_player},{position_player},{age},{pace},{Shooting},{passing},{Dribbling},{Defending},{Physical},{Attacking_Work_Rate},{Defensive_Work_Rate},{Preferred_Foot},{club},{player_nationality},{Player_height},{Player_weight},{birthday_of_player}\n")
            except  :
                print("i am tired....")
```
